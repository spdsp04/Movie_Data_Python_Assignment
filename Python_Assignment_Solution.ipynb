```python

# Importing Pandas library and assigning it the alias 'pd'
import pandas as pd

# Reading the data-set as a DataFrame
movies = pd.read_csv('Movie Assignment Data.csv')
# Displaying the first 5 rows of the DataFrame
movies.head()

```

## Question 2: Generate descriptive statistics for the budget of all the movies.
    Solution:
        
        - A descriptive statistical analysis can be done using describe()
        - Additionally analysis can be derived using
            - Use the count() function to calculate the count of movies produced
            - Use the sum() function to calculate the total budget of all movies produced
            - Use the max() function to calculate the budget of the most expensive movie produced
            - Use the min() function to calculate the budget of the least expensive movie produced
            - Use the mean() function to calculate the average budget of all movies produced
            - Use the median() function to calculate the median of all budgets
            - Use the std() function to calculate the standard deviation of budgets
            - Use the quantile() function to calculate the 25th, 50th and 75th quantile
        


```python
# Creating a series of all budgets
budget = movies['budget']

# A descriptive statistical analysis can be done using describe()
print(budget.describe())

print(f"Count of movie produced: {budget.count()}") # Count of movie produced using count()
print(f"Total budget of all movies produced: {budget.sum()}") # Total budget of all movies produced using sum()
print(f"Most expensive movie produced had a budget of: {budget.max()}") # Most expensive movie using max()
print(f"Least expensive movie produced had a budget of: {budget.min()}") # Least expensive movie using min()
print(f'Average budget of all movies produced: {budget.mean()}') # Average budget using mean()
print(f'Median of all budgets: {budget.median()}') # Median budget using median()
print(f'Standard deviation of all budgets: {budget.std():.2f}') # Standard deviation using std()
print(f'25th quantile: {budget.quantile(q=0.25)}') # 25th quantile using quantile()
print(f'50th quantile: {budget.quantile(q=0.50)}') # 50th quantile using quantile()
print(f'75th quantile: {budget.quantile(q=0.75)}') # 75th quantile using quantile()

```

## Question 3: Find out how many of the top-rated movies produced in the United States have a PG-13 rating
    Solution:
        - Filtering using column 'content_rating' = 'PG-13': movies['content_rating'] == 'PG-13'
        - Filtering using column 'Country' = 'USA': movies['Country'] == 'USA'
        - Using & operator to match both conditions
        - Using len() to find the count of rows returned

```python
# To find out how many movies were produced in the USA with a PG-13 rating we will be filtering the Dataframe by content_rating and Country.
# Then using the len() function count the number of movies

print(f"Number of movies that were produced in the USA with PG-13 rating are: \
{len(movies[(movies['content_rating'] == 'PG-13') & (movies['Country'] == 'USA')])}")

```

## Question 4: Find out whether any of the top-rated movies produced in 2014 were not produced in the United States
    Solution:
        - Filtering using column 'title_year' = 2014: movies['title_year'] == 2014
        - Filtering using column 'Country' = 'USA': movies['Country'] != 'USA'
        - Using & operator to match both conditions
        - Filter movies based on conditions and store it in 'non_USA_Movies'
        - Using len() to find the count of rows returned

```python
# Filtering using title_year and Country we can get the list of movies that were produced in 2014 but not in the USA

# Filtering movies based on filter
non_USA_Movies = movies[(movies['title_year'] == 2014) & (movies['Country'] != 'USA')] 

# Using len() to find the count of rows
print(f'There were {len(non_USA_Movies)} movies that were produced in 2014 but not in the USA.')

```

## Question 5: Find the percentage of the top-rated movies that are in:
    - 1 genre only
    - 2 genres only
    - 3 genres only

    Solution:
        - Create condition filter 'condition_genre_3_null' for all movies where 'genre_3' is NaN
        - Create condition filter 'condition_genre_2_null' for all movies where 'genre_2' is NaN
        - Create condition filter 'condition_genre_2_not_null' for all movies where 'genre_2' is not null
        - Create condition filter 'condition_genre_3_not_null' for all movies where 'genre_3' is not null
        - Use len() to count number of values returned
        - Use round() to round the value to zeroth value

```python
# Defining filter conditions

condition_genre_3_null = movies['genre_3'].isnull() # For all movies that have 'genre_3' as NaN
condition_genre_2_null = movies['genre_2'].isnull() # For all movies that have 'genre_2' as NaN
condition_genre_2_not_null = movies['genre_2'].notna() # For all movies that have value in 'genre_2'
condition_genre_3_not_null = movies['genre_3'].notna() # For all movies that have value in 'genre_3'

# By filtering using conditions 'genre_3' == NaN and 'genre_2' == NaN, we can get movies that have only 1 genre
# Get the count of movies using len(movies)
# Divide the count of movies returned using condition filter by count of all movies
# Round the value to the zeroth value using round()
print(f'Percentage of top-rated movies that are in 1 genre only: {round((len(movies[condition_genre_3_null & condition_genre_2_null])/len(movies))*100)}%')

# By filtering using conditions 'genre_3' == NaN and 'genre_2' has value, we can get movies that have only 2 genre
# Get the count of movies using len(movies)
# Divide the count of movies returned using condition filter by count of all movies
# Round the value to the zeroth value using round()
print(f'Percentage of top-rated movies that are in 2 genre only: {round((len(movies[condition_genre_3_null & condition_genre_2_not_null])/len(movies))*100)}%')

# By filtering using condition 'genre_3' has value, we can get movies that have all 3 genre
# Get the count of movies using len(movies)
# Divide the count of movies returned using condition filter by count of all movies
# Round the value to the zeroth value using round()
print(f'Percentage of top-rated movies that are in 3 genre only: {round((len(movies[condition_genre_3_not_null])/len(movies))*100)}%')

```

## Question 6: Convert the budget and gross values from “dollars” to “dollars in millions” for all top-rated movies. Round the converted values down to 3 decimal places. For example, a value of 192,345,273 should be converted to 192.345
    Solution:
        - Get the 'dollors in millions' value by dividing 'budget' by 1000,000
        - Use round() to round the value to 3 decimal places

```python
# To convert the 'bugdet' and 'Gross' value from dollors to dollors in million we will have to divide by 1000000
# and to round the result to 3 decimal places we can use round()

budget_in_millions = round((movies['budget'] / 1000000), 3) # Converting budget values to millions by dividing by 100,000 and rounding to 3 decimal places
movies['budget (millions)'] = budget_in_millions # Adding a new column named 'budget (millions)' to the Dataframe

gross_in_millions = round((movies['Gross'] / 1000000), 3) # Converting gross values to millions by dividing by 100,000 and rounding to 3 decimal places
movies['Gross (millions)'] = gross_in_millions # Adding a new column named 'Gross (millions)' to the Dataframe

filter = ['Title', 'title_year', 'budget', 'budget (millions)', 'Gross', 'Gross (millions)'] # Creating filter to display selected columns
movies[filter] # Displaying Dataframe with added columns

```

## Question 7:  List all details for the top 10 movies with the highest profit, sorted from highest to lowest. Hint: Profit is not a column in the DataFrame. You will need to calculate it
    Solution:
        - Calculate the profit by substracting the 'budget' from 'Gross'
        - Create a new column name 'Profit' and add the calculated profit
        - Use sort_values() to sort by 'Profit' in descending order
        - Use head() to display the first 10 rows

```python
# To calculate the profit earned by each movie, we will substract the 'budget' from 'Gross'
# We will create a new column in the DataFrame named 'Profit' and add the calculated profit
movies['Profit'] = movies['Gross'] - movies['budget']

# Finally we will sort the DataFrame in descending order by 'Profit' and select the first 10 rows
movies.sort_values(by='Profit', ascending=False).head(10)

```

## Question 8: Generate a list of all the actors, in alphabetical order by the first name, that have starred in a top-rated movie. If an actor has starred in multiple movies, their name should appear only once on the list. Assume that all actors’ names are in the format <first_name> <last_name>
    Solution:
        - Create the list of names of actors in column 'actor_1_name'
        - Create the list of names of actors in column 'actor_2_name'
        - Create the list of names of actors in column 'actor_3_name'
        - Use pd.concat() to create a list of all names in the above created 3 lists of names
        - Use unique() to filter out duplicate names
        - Use sorted() to sort the names in alphabetical order

```python
# 

actor_1_name = movies['actor_1_name'] # Getting all actors in column 'actor_1_name'
actor_2_name = movies['actor_2_name'] # Getting all actors in column 'actor_2_name'
actor_3_name = movies['actor_3_name'] # Getting all actors in column 'actor_3_name'
all_actors = pd.concat([actor_1_name, actor_2_name, actor_3_name]) # Creating an all_actors list by concatinating the names
unique_all_actors = all_actors.unique() # Getting all unique names
print(sorted(unique_all_actors)) # Printing the actors' name in ascending order

```

## Question 10: Create a data visualization that shows each country and the number of top-rated movies produced in it. Find the country that produced the most top-rated movies
    Solution:
        - Import pyplot from matplotlib as plt
        - Use the value_counts() to get the count of movies per country
        - Convert the series object to a DataFrame
        - Create frequency chart
        - Plot the graph

<div class="alert alert-block alert-warning">
<b>Plot attributes:</b> 
    <ol>
        <li><strong>Type</strong>: Bar Chart</li>
        <li><strong>Figure Size</strong>: 10 x 6 inches</li>
        <li><strong>Title</strong>: Number of movies produced by country</li>
        <li><strong>X-axis</strong>: Country</li>
        <li><strong>Y-axis</strong>: Count</li>
    </ol>
</div>

```python
# Importing matplotlib to plot the bar chart
import matplotlib.pyplot as plt

# Getting the count of movies per country using value_counts() and converting the Series object returned to a Dataframe
movies_countries = movies['Country'].value_counts().to_frame()
# Plotting the graph
# Defining the type of the graph to be plotted as 'bar' and size of the plot as 10x 6
movies_countries.plot(kind='bar', y='Country', figsize=(10, 6))
plt.title('Number of movies produced by country') # Labeling the title of the graph
plt.xlabel('Country') # Defining X-axis as 'Country'
plt.ylabel('Count') # Defining Y-axis as Count
plt.show() # Displaying the graph

```
